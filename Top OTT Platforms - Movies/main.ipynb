```python
import pandas as pd
data3 = pd.read_csv(r"movies.csv")
data3.head()
```

```python
mrating_count = {'Netflix': {'9': 0, '8': 0, '7': 0, '6': 0, '5': 0, '4': 0, '3': 0, '2': 0, '1': 0, 'total': 0}, 
                'Hulu': {'9': 0, '8': 0, '7': 0, '6': 0, '5': 0, '4': 0, '3': 0, '2': 0, '1': 0, 'total': 0},
                'Prime': {'9': 0, '8': 0, '7': 0, '6': 0, '5': 0, '4': 0, '3': 0, '2': 0, '1': 0, 'total': 0},
                'Disney+': {'9': 0, '8': 0, '7': 0, '6': 0, '5': 0, '4': 0, '3': 0, '2': 0, '1': 0, 'total': 0}}

for i in range(data3.shape[0]):
  try:
    rating = data3.iloc[i][4][0]
  except:
    continue
  if data3.iloc[i][5]:
    mrating_count['Netflix'][rating] += 1
    mrating_count['Netflix']['total'] += 1
  if data3.iloc[i][6]:
    mrating_count['Hulu'][rating] += 1
    mrating_count['Hulu']['total'] += 1
  if data3.iloc[i][7]:
    mrating_count['Prime'][rating] += 1
    mrating_count['Prime']['total'] += 1
  if data3.iloc[i][8]:
    mrating_count['Disney+'][rating] += 1
    mrating_count['Disney+']['total'] += 1

print(mrating_count['Netflix'])
print(mrating_count['Hulu'])
print(mrating_count['Prime'])
print(mrating_count['Disney+'])
```

```python
for i in ['Netflix', 'Prime', 'Hulu', 'Disney+']:
  print(i)
  for k,v in mrating_count[i].items():
    print(k, round(v/mrating_count[i]['total']*100, 2))
  print()
```
